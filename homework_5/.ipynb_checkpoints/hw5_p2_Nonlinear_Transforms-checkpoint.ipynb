## Problem 2

We are given a plot of the box $[-1,1] \times [-1,1]$ with two hyperbolas which separate the box into three regions. Points in the left and right region are classified as $-1$, the points in the center are classified as $+1$. 

```python
import numpy as np
import matplotlib.pyplot as plt


def classify(X, w):
    return np.sign(np.dot(X, w))


def problem2():
    
    w = np.array([0.2, -2, 1])
    
    
    N = 1000
    
    x1 = np.random.uniform(-1,1,N)
    x2 = np.random.uniform(-1,1,N)
    X = np.array([np.ones(N), x1*x1, x2*x2]).T
    
    classification = classify(X, w)
    
    plt.plot(x1[classification==-1], x2[classification==-1], 'r_')
    plt.plot(x1[classification==+1], x2[classification==+1], 'b+')
    
    plt.ylim(-1,1)
    plt.xlim(-1,1)
    plt.show()



problem2()
```
